```python
COVIGraph = go.Figure(data=go.Choropleth(
    locations=COVIData['state_abrv'], # Spatial coordinates
    z = COVIData["covIndex"].astype(float), # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = "Reds",
    colorbar_title = "Cost of Voting",
))
COVIGraph.update_layout(
    title_text = '2020 Cost of Voting Index',
    geo_scope='usa', # limite map scope to USA
)
COVIGraph.show()
```
