```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('COVdata.csv')
```

```python
df.head()
```

```python
df['regScore'] = np.nan
```

```python
df['registration deadline']=pd.to_datetime(df['registration deadline'])
df['starting day of voting early']=pd.to_datetime(df['starting day of voting early'])
df['deadline to request']=pd.to_datetime(df['deadline to request'])
df['date to return absentee by']=pd.to_datetime(df['date to return absentee by'])
```

```python
df.regScore = (pd.to_datetime('2020-11-03')-df['registration deadline']).dt.total_seconds()/(60*60*24)
df.regScore=df.regScore/df.regScore.max()
# Calculates the index for voter registration by finding the difference between the state's
# deadline and November 3rd, which is the latest date of any state. It then finds the ratio of 
# the difference in time between that state's registration date and november 3rd with that of the 
# state with the earliest registration deadline. Large -> earlier registration, and less 
# accomodating to voters.
```

```python
df.head()
```

```python
df['earlyVoteScore'] = np.nan
```

```python
df.earlyVoteScore=(df['starting day of voting early']-df['starting day of voting early'].min()).dt.total_seconds()/(60*60*24)
df.earlyVoteScore=df.earlyVoteScore/df.earlyVoteScore.max()
df.head()
# Calculates the index for early voting dates
# Finds ratio of time from earliest start for early voting to max time difference from earliest start of any state
# High index means later early voting start, and harder to vote
```

```python
df['idScore'] = df['id for in person voting']/2
```

```python
df['requestScore'] = df['who can request']/2
```

```python
df['absenteeMailScore'] = df['absentee mailed out']/2
```

```python
df['witnessScore'] = df['witness needed']
```

```python
df.head()
```

```python
df['requestDateScore']=(df['deadline to request']-df['deadline to request'].min()).dt.total_seconds()/(60*60*24)
df.requestDateScore=1-(df.requestDateScore/df.requestDateScore.max())
```

```python
df['returnScore']=(df['date to return absentee by']-df['date to return absentee by'].min()).dt.total_seconds()/(60*60*24)
df.returnScore=1-(df.returnScore/df.returnScore.max())
```

```python
df['whoRequestScore']=df['who can request']/2
```

```python
df['covIndex']=((df.regScore+df.earlyVoteScore + df.idScore+ df.requestScore+df.absenteeMailScore + df.witnessScore+df.returnScore+df.whoRequestScore)/2)-2
```
